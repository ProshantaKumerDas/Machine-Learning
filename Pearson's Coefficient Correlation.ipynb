```python
import numpy as np
length = len(a)
print(length)

mean_x = np.mean(a)
mean_y = np.mean(b)

print("Mean of a: ",mean_x)
print("Mean of b: ",mean_y)
```

```python
a1=0
b1=0
c1=0
for i in range(length):
 
  a1 = a1+ np.sum((a[i] - mean_x) * (b[i] - mean_y))
  b1 = b1 + np.sum((a[i]-mean_x)**2)
  c1 = c1 + np.sum((b[i]-mean_y)**2)

b1sq = np.sqrt(b1)
c1sq = np.sqrt(c1)
r = a1/(b1sq*c1sq)
print("Pearson's Coefficient r= ", r)
```
